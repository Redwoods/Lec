### Graph of all signals

```r
par(mfrow=c(3,1))  # without noise
plot(s_all$temperature, pch=20, main="IOT temperature", xlab="", col=2)
plot(s_all$humidity, pch=20, main="IOT humidity", xlab="", col=3)
plot(s_all$luminosity, pch=20, main="IOT luminosity", xlab="data index", col=4)
```
