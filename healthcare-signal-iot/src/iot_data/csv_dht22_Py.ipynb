### Convert dataframe to numpy array

```python
# Convert dataframe to numpy array
d3a = d3.values  
```

### Multiple histogram

```python
n_bins = 20
# plt.figure(figsize=(12,7))
fig, axes = plt.subplots(2,2,figsize=(9,9))  # subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = axes.flatten()

# colors = ['red', 'tan', 'lime']
ax1.hist(data_all.temperature, n_bins)  
# ax1.legend(prop={'size': 10})
ax1.set_title('Temperature')

ax2.hist(data_all.humidity, n_bins)  
# ax1.legend(prop={'size': 10})
ax2.set_title('Humidity')

ax3.hist(data_all.luminosity, n_bins) 
# ax1.legend(prop={'size': 10})
ax3.set_title('Luminosity')

colors = ['red', 'tan', 'lime']
labels=['Temp', 'Humi', 'Lumi']
ax4.hist(d3a, n_bins, histtype='bar', color=colors, label=labels)
ax4.legend(prop={'size': 10})
ax4.set_title('IOT histogram')

fig.tight_layout()
plt.show()
```
