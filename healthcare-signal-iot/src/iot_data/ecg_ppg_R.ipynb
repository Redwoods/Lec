### Load EEG & PPG data (ecg_ppg_s1.csv)

```r
ecg_ppg <- read.csv("data/ecg_ppg_s1.csv", header=T)
```

```r
head(ecg_ppg)
```

```r
summary(ecg_ppg)
```

```r
str(ecg_ppg)
```

### Graph of all data

```r
par(mfrow=c(2,1)) 
plot(ecg_ppg$ecg, type='l', pch=20, main="ECG", xlab="", col=2, lwd=1)
plot(ecg_ppg$ppg, type='l', pch=20, main="PPG", xlab="data index", col=3, lwd=1)
```

```r
# 시간 환산
dim(ecg_ppg)[1]*5/1000  # fs=200 Hz
# sec
```

```r
par(mfrow=c(2,1))  # graph over time
plot((1:dim(ecg_ppg)[1])*5/1000, ecg_ppg$ecg, type='l', pch=20, main="ECG", xlab="sec", col=2, lwd=1)
plot((1:dim(ecg_ppg)[1])*5/1000, ecg_ppg$ppg, type='l', pch=20, main="PPG", xlab="sec", col=3, lwd=1)
```

### Graph of initial 10 waves of ECG & PPG

```r
# ECG & PPG (Initial 10 waves)
ep10 <- ecg_ppg[10:1620,]
par(mfrow=c(2,1))  
plot(ep10$ecg, type='l', pch=20, main="ECG", xlab="data index", col=2, lwd=2)
plot(ep10$ppg, type='l', pch=20, main="PPG", xlab="data index", col=3, lwd=2)
plot((10:1620)*5/1000, ep10$ecg, type='l', pch=20, main="ECG", xlab="sec", col=2, lwd=2)
plot((10:1620)*5/1000, ep10$ppg, type='l', pch=20, main="PPG", xlab="sec", col=3, lwd=2)
```

### histogram

```r
par(mfrow=c(1,2))
hist(ecg_ppg$ecg, col="red")
hist(ecg_ppg$ppg, col="green")
```

### Load HRV & PRV data (hrv_prv_s1.csv)

```r
hrv_prv <- read.csv("data/hrv_prv_s1.csv", header=T)
```

```r
summary(hrv_prv)
```

```r
str(hrv_prv)
```

```r
head(hrv_prv)
```

### Graph of HRV & PRV

```r
# HRV & PRV
par(mfrow=c(2,1))  
plot(hrv_prv$HRV, type='l', pch=20, main="HRV (bpm)", xlab="beat index", col=2, lwd=2)
plot(hrv_prv$PRV, type='l', pch=20, main="PRV (bpm)", xlab="beat index", col=3, lwd=2)
plot(cumsum(hrv_prv$HPV)/1000, hrv_prv$HRV, type='l', pch=20, main="HRV (bpm)", xlab="sec", col=2, lwd=2)
plot(cumsum(hrv_prv$PPV)/1000, hrv_prv$PRV, type='l', pch=20, main="PRV (bpm)", xlab="sec", col=3, lwd=2)
```

```r
# HRV & PRV (initial 50 data)
par(mfrow=c(2,1))  
plot(hrv_prv$HRV[1:50], type='b', pch=20, main="HRV (bpm)", xlab="beat index", col=2, lwd=2)
plot(hrv_prv$PRV[1:50], type='b', pch=20, main="PRV (bpm)", xlab="beat index", col=3, lwd=2)
plot(cumsum(hrv_prv$HPV[1:50])/1000, hrv_prv$HRV[1:50], type='b', pch=20, main="HRV (bpm)", xlab="sec", col=2, lwd=2)
plot(cumsum(hrv_prv$PPV[1:50])/1000, hrv_prv$PRV[1:50], type='b', pch=20, main="PRV (bpm)", xlab="sec", col=3, lwd=2)
```

### Histogram of heart rate & pulse rate

```r
par(mfrow=c(1,2))
hist(hrv_prv$HRV, col="red", xlab="HR (bpm)")
hist(hrv_prv$PRV, col="green", xlab="PR (bpm)")
```
